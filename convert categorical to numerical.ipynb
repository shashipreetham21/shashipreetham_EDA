**one heart encoder**

- one heart encoder name says at a time one will ON and other is OFF

- for example: case status has two labels one is certified and one is denied
    
- so when you apply one heart encoding it creates twom extra columns that is
 
     - case_status_certified
        
     - case_status_denied 

**advantage**

- when you develop ML model it is very important the columns should be independent each other

- so here case status creating two extra columns

- which are independent each other,which menas the row values at a time only one column has 1

- 

**disadvantages**

- the disadvantage is if a column has 100 unique values,100 new columns will be there

- the data will become spares, which means huge

- columns are more means, dimensions are more

- the processing time is more

- the memory consumption is more

- curse of dimensionality(more columns)
